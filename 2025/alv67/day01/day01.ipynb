# --- Day 1: Secret Entrance ---


```python
# --- Support Functions ---
# This section contains support functions used by the main code.

def read_input(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()
```

```python
# --- Part ONE ---
input = read_input('input.txt')

DIAL_START = 50
MAX_DIAL = 100

dial = DIAL_START
count_zeros = 0

for rot in input:
    direction = rot[0]
    amount = int(rot[1:])

    dial = ((dial + amount) if direction == 'R' else (dial - amount)) % MAX_DIAL
    if dial == 0:
        count_zeros += 1

result = count_zeros
print("Part ONE:", result)
```

```python
# --- Part TWO ---
input = read_input('input.txt')

DIAL_START = 50
MAX_DIAL = 100

dial = DIAL_START
count_zeros = 0

for rot in input:
    direction = rot[0]
    amount = int(rot[1:])
    zeroes = amount // MAX_DIAL
    amount = amount % MAX_DIAL

    if amount:
        if direction == 'R':
            dial += amount
            zeroes += 1 if dial > MAX_DIAL else 0
        else:
            dial -= amount
            zeroes += 1 if dial < 0 and dial != -amount else 0
        dial = dial % MAX_DIAL
        if dial == 0:
            zeroes += 1
        count_zeros += zeroes
    # print (f" {rot:6} => Dial: {dial:<6}, Zeroes: {zeroes}")

result = count_zeros
print("Part TWO:", result)
```
