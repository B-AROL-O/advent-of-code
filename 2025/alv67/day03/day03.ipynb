```python
# --- Support Functions ---
# This section contains support functions used by the main code.

def read_input(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()
```

```python
# --- Part ONE ---
input = read_input('input.txt')

#print (input)

result = 0
for bank in input:
    bank = [int(x) for x in bank]
    #print (bank)
    d1 = max(bank[:-1])
    d2 = max(bank[bank.index(d1)+1:])
    #print (d1, d2)
    result += d1*10 + d2

print("Part ONE:", result)
```

```python
# --- Part TWO ---
input = read_input('input.txt')

NR_DIGITS = 12

result = 0
for bank in input:
    bank = [int(x) for x in bank]
    voltage = 0
    idx = 0
    for digit in range(1,NR_DIGITS+1):
        ridx = len(bank)-(NR_DIGITS-digit)
        s = bank[idx:ridx]
        val = max(s)
        idx = bank.index(val,idx)+1
        voltage = voltage*10 + val
    result += voltage

print("Part TWO:", result)
```
