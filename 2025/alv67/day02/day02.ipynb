# --- Day 2: Gift Shop ---


```python
# --- Support Functions ---
# This section contains support functions used by the main code.

def read_input(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()
```

```python
# --- Part ONE ---
input = read_input('input.txt')

ids = input[0].split(',')

# print (ids)

result = 0

for id in ids:
    # print (id, end=' -> ')
    a, b = id.split('-')
    for x in range(int(a), int(b)+1):
        x = str(x)
        # print (x, end=' ')
        if x[:len(x)//2] == x[len(x)//2:]:
            # print (x, end=' ')
            result += int(x)
    # print ("")


result = result
print("Part ONE:", result)
```

```python
# --- Part TWO ---
input = read_input('input.txt')

ids = input[0].split(',')

# print (ids)

result = 0

for id in ids:
    # print (id, end=' -> ')
    a, b = id.split('-')
    for x in range(int(a), int(b)+1):
        x = str(x)
        for parti in range(2, len(x)+1):
            # print (f"parti={parti}", end=' ')
            if len(x) % parti != 0:
                continue
            l = len(x) // parti
            parts = [x[i*l:(i+1)*l] for i in range(parti)]
            
            if all(part == parts[0] for part in parts):
                # print (x, end=' ')
                result += int(x)
                break
    # print ("")

result = result
print("Part TWO:", result)
```
